# Model Training

```python
# Use GPU if possible
if torch.cuda.is_available():      
    device = torch.device("cuda")
    print('There are %d GPU(s) available.' % torch.cuda.device_count())
    print('Using GPU:', torch.cuda.get_device_name(0))
else:
    device = torch.device("cpu")
    print('No GPU available, using the CPU instead.')
```

```python
# initialize model
model = SpecAutoEncoder()
model_name = "SpecAutoEncoder"
```

```python
# set hyperparams
learning_rate = 1e-3
optimizer = torch.optim.AdamW(model.parameters(), lr=learning_rate, eps=1e-8)
criterion = MSELoss()
batch_size = 128
epochs = 32

# Set random seeds.
random_seed = 123
np.random.seed(random_seed)
torch.manual_seed(random_seed)
torch.cuda.manual_seed(random_seed)
random.seed(random_seed)
```

```python
# load datasets
train_dataset = torch.load("../data/datasets/spec/train.pt")
valid_dataset = torch.load("../data/datasets/spec/valid.pt")
test_dataset = torch.load("../data/datasets/spec/test.pt")

# create dataloaders
train_dataloader = DataLoader(train_dataset, batch_size=batch_size)
valid_dataloader = DataLoader(valid_dataset, batch_size=batch_size)
test_dataloader = DataLoader(test_dataset, batch_size=batch_size)
```

```python
# initialize early stopper
early_stopper = EarlyStopper(patience=5, delta=0., output_dir="../models")
```

```python
# train the model
train_stats = train(model, criterion, optimizer, epochs, train_dataloader, valid_dataloader, early_stopper, device, model_name)
```

```python
# plot the training & validation loss
plot_train_val_loss(train_stats=train_stats, figure_path=f"../figures/{model_name}_train_val_loss.png")
```
