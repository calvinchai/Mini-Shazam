```python
import pandas as pd
```

```python
rl=pd.read_csv('rl_record.csv', index_col=0)
cp=pd.read_csv('chromaprint_record.csv', index_col=0)
cqt=pd.read_csv('cqt_record.csv', index_col=0)
```

```python
#plot the recall 1,5,10 for each dataset
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
width = 0.2
"""
Frequency-Domain Autoencoder   &   0.09   &  0.125   &  0.14625  \\
        Time-Domain Autoencoder        &   0.03   &  0.0475  &  0.0575   \\
            """
y1 = [0.09, 0.125, 0.14625]
y2 = [0.03, 0.0475, 0.0575]
y3 = [cp['recall1'].mean(), cp['recall5'].mean(), cp['recall10'].mean()]
y4 = [cqt['recall1'].mean(), cqt['recall5'].mean(), cqt['recall10'].mean()]

fig, ax = plt.subplots()


plt.plot(x, y1, label='Time-Domain Autoencoder')
plt.plot(x, y2, label='Frequency-Domain Autoencoder')
plt.plot(x, y3, label='Chromaprint')
plt.plot(x, y4, label='CQT', linestyle='--')

ax.set_ylabel('Recall')
ax.set_xlabel('Recall @ N')
ax.set_title('Recall with whole track')
ax.set_xticks(x)
ax.set_xticklabels(['1', '5', '10'])
ax.legend()

plt.show()

```

```python
# for each dataset, plot the average first 1,5,10 lookup

x = np.arange(3)
width = 0.2
y1 = [rl['lookup'][:1].mean(), rl['lookup'][:5].mean(), rl['lookup'][:10].mean()]
y2 = [cp['lookup'][:1].mean(), cp['lookup'][:5].mean(), cp['lookup'][:10].mean()]
y3 = [cqt['lookup'][:1].mean(), cqt['lookup'][:5].mean(), cqt['lookup'][:10].mean()]

fig, ax = plt.subplots()
plt.plot(x, y1, label='RL')
plt.plot(x, y2, label='Chromaprint')
plt.plot(x, y3, label='CQT', linestyle='--')

ax.set_ylabel('Runtime')
ax.set_xlabel('First N')
ax.set_title('Average Runtime (s) for first N lookup')
ax.set_xticks(x)
ax.set_xticklabels(['1', '5', '10'])
ax.legend()

plt.show()
```
